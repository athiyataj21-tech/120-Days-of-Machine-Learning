```python
student = {
    "name": "Athiya",
    "age": 22,
    "course": "Machine Learning"
}

student
```

```python
student["name"], student["course"]
```

```python
student.get("age"), student.get("city")
```

```python
student["city"] = "Bangalore"

student
```

```python
student["age"] = 23

student
```

```python
student.update({"college": "Government Engineering College, Hassan"})

student
```

```python
print(student.keys())
print(student.values())
print(student.items())
```

```python
for key in student:
    print(key)
```

```python
for key, value in student.items():
    print(key, value)
```

```python
"name" in student, "salary" in student
```

```python
student.pop("city")

student
```

```python
students = {
    101: {"name": "Alice", "marks": 92},
    102: {"name": "Bob", "marks": 88},
    103: {"name": "Charlie", "marks": 95}
}

students
```

```python
students[101]["name"], students[103]["marks"]
```

```python
square = {x: x**2 for x in range(1,11)}

square
```

```python
dict1 = {"a":1, "b":2}
dict2 = {"c":3, "d":4}

dict1.update(dict2)

dict1
```

```python
text = "machinelearning"

frequency = {}

for char in text:
    frequency[char] = frequency.get(char, 0) + 1

frequency
```

```python
marks = {
    "Math":95,
    "Python":98,
    "Machine Learning":93,
    "Statistics":90
}

print(min(marks.values()))
print(max(marks.values()))
print(sum(marks.values()))
print(len(marks))
```

```python
person = {
    "name":"Athiya",
    "age":22,
    "city":"Bangalore"
}

print(person)
```

```python
employee = {
    "id":101,
    "name":"John",
    "department":"Data Science",
    "salary":70000
}

for key, value in employee.items():
    print(key, value)
```

```python
cube = {x:x**3 for x in range(1,11)}

cube
```
