# Day 06 - Python Lists

Learn Python lists with examples and exercises.

## Creating Lists

```python
fruits=['Apple','Banana','Orange']
numbers=[10,20,30]
print(fruits)
print(numbers)
```

## Accessing Elements

```python
fruits=['Apple','Banana','Orange','Mango']
print(fruits[0])
print(fruits[-1])
```

## Slicing

```python
fruits=['Apple','Banana','Orange','Mango','Kiwi']
print(fruits[1:4])
print(fruits[::-1])
```

## Updating

```python
fruits=['Apple','Banana']
fruits[1]='Mango'
print(fruits)
```

## Adding Elements

```python
fruits=['Apple']
fruits.append('Banana')
fruits.insert(1,'Kiwi')
fruits.extend(['Orange','Grapes'])
print(fruits)
```

## Removing Elements

```python
fruits=['Apple','Banana','Orange']
fruits.remove('Banana')
print(fruits)
print(fruits.pop())
```

## Built-in Functions

```python
nums=[5,2,9,1]
print(len(nums),sum(nums),max(nums),min(nums))
```

## Sorting

```python
nums=[5,2,9,1]
nums.sort()
print(nums)
nums.reverse()
print(nums)
```

## Nested Lists

```python
matrix=[[1,2],[3,4]]
print(matrix[0][1])
```

## Loops

```python
for f in ['Apple','Banana','Orange']:
    print(f)
```

## List Comprehension

```python
squares=[x*x for x in range(6)]
print(squares)
even=[x for x in range(11) if x%2==0]
print(even)
```

## Exercise

```python
marks=[85,90,78,92,88]
print('Total',sum(marks))
print('Average',sum(marks)/len(marks))
print('Highest',max(marks))
print('Lowest',min(marks))
```
