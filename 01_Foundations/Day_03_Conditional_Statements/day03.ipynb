# Day 3 - Conditional Statements in Python

Topics Covered

- if Statement
- if...else
- if...elif...else
- Nested if
- Comparison Operators
- Logical Operators
- Membership Operators
- Identity Operators
- Ternary Operator

```python
age = 20

if age >= 18:
    print("Eligible to vote")
```

```python
number = 15

if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
marks = 82

if marks >= 90:
    print("A Grade")
elif marks >= 75:
    print("B Grade")
elif marks >= 60:
    print("C Grade")
else:
    print("Fail")
```

```python
age = 22
citizen = True

if age >= 18:
    if citizen:
        print("Eligible")
    else:
        print("Citizen Required")
else:
    print("Underage")
```

```python
username = "admin"
password = "1234"

if username == "admin" and password == "1234":
    print("Login Successful")
else:
    print("Invalid Login")
```

```python
num = 25

result = "Positive" if num > 0 else "Negative"

print(result)
```

```python
marks = int(input("Enter Marks: "))

if marks >= 90:
    print("Excellent")
elif marks >= 75:
    print("Very Good")
elif marks >= 60:
    print("Good")
elif marks >= 35:
    print("Pass")
else:
    print("Fail")
```
