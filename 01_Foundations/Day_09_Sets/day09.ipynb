# What is a Set?

A **Set** is one of Python's built-in data structures used to store multiple items in a single variable.

Unlike Lists and Tuples,

- Sets store **only unique values**
- Duplicate values are removed automatically
- Sets are unordered
- Sets are mutable
- Sets are very fast because they use **Hash Tables**

They are mainly used when:

- Removing duplicate values
- Finding common elements
- Performing mathematical operations
- Fast searching

````python
fruits = {"Apple", "Banana", "Orange"}

print(fruits)
````

## Why Use Sets?

Sets are useful in many real-world applications.

Examples:

- Removing duplicate records
- Comparing two datasets
- Finding common friends
- Machine Learning preprocessing
- Recommendation systems
- Database operations

Since Sets are implemented using Hash Tables, searching is much faster than Lists.

````python
numbers = [1,2,2,3,4,5,5,6]

unique_numbers = set(numbers)

print(unique_numbers)
````

# Creating Sets

There are several ways to create Sets in Python.

## Method 1

Using Curly Braces {}

This is the most common method.

````python
colors = {"Red", "Green", "Blue"}

print(colors)
````

## Method 2

Using the set() constructor.

````python
numbers = set([10,20,30,40])

print(numbers)
````

## Method 3

Creating a Set from a String.

Duplicate characters are automatically removed.

````python
letters = set("Machine")

print(letters)
````

# Empty Set

Many beginners make a common mistake while creating an empty Set.

Using `{}` creates an empty Dictionary, **not** an empty Set.

To create an empty Set, use the `set()` function.

````python
A = {}

print(type(A))
````

````python
B = set()

print(type(B))
````

A Set can contain:

- Integers
- Float

# Accessing Set Elements

Unlike Lists, Sets do not support indexing.

Instead, we iterate through Sets using loops.

````python
fruits = {"Apple","Banana","Orange"}

for fruit in fruits:
    print(fruit)
````

````python
A = {10,20,30}

print(A[0])
````

# Membership Operators

Python provides two membership operators for Sets.

- `in`
- `not in`

Since Sets use Hash Tables internally, membership testing is extremely fast (average O(1) time complexity).

Membership operators are commonly used to:

- Check whether an element exists
- Validate user input
- Remove duplicates
- Search large datasets efficiently

````python
fruits = {"Apple", "Banana", "Orange"}

print("Apple" in fruits)
print("Mango" in fruits)
````

````python
numbers = {10,20,30,40}

print(50 not in numbers)
print(20 not in numbers)
````

# Adding Elements

The `add()` method inserts a single element into a Set.

### Syntax

```python
set_name.add(element)
```

If the element already exists, nothing changes because Sets only store unique values.

````python
fruits = {"Apple", "Banana"}

fruits.add("Orange")

print(fruits)
````

````python
numbers = {1,2,3}

numbers.add(4)

print(numbers)
````

````python
numbers = {1,2,3}

numbers.add(2)

print(numbers)
````

# Updating a Set

The `update()` method adds multiple elements to a Set.

Unlike `add()`, which inserts only one item, `update()` can insert:

- Lists
- Tuples
- Sets
- Strings
- Multiple Iterables

### Syntax

```python
set_name.update(iterable)
```

````python
fruits = {"Apple","Banana"}

fruits.update(["Orange","Mango"])

print(fruits)
````

````python
A = {1,2,3}

B = {4,5,6}

A.update(B)

print(A)
````

````python
letters = {"A","B"}

letters.update("CD")

print(letters)
````

# Removing Elements

Python provides several methods to remove elements from a Set.

- remove()
- discard()
- pop()
- clear()

Each method behaves differently.

## remove()

The `remove()` method deletes a specified element.

If the element does not exist, Python raises a **KeyError**.

### Syntax

```python
set_name.remove(element)
```

````python
fruits = {"Apple","Banana","Orange"}

fruits.remove("Banana")

print(fruits)
````

````python
fruits = {"Apple","Banana"}

fruits.remove("Mango")
````

## discard()

The `discard()` method also removes an element.

Unlike `remove()`, it does **not** raise an error if the element is absent.

### Syntax

```python
set_name.discard(element)
```

````python
fruits = {"Apple","Banana"}

fruits.discard("Banana")

print(fruits)
````

````python
fruits = {"Apple","Banana"}

fruits.discard("Mango")

print(fruits)
````

# remove() vs discard()

| remove() | discard() |
|-----------|-----------|
| Raises KeyError | No Error |
| Deletes existing element | Deletes existing element |
| Fails if absent | Ignores if absent |

# pop()

The `pop()` method removes and returns a random element from the Set.

Since Sets are unordered, you cannot predict which element will be removed.

````python
colors = {"Red","Green","Blue"}

removed = colors.pop()

print("Removed:", removed)
print("Remaining:", colors)
````

# clear()

The `clear()` method removes all elements from the Set while keeping the Set object itself.

### Syntax

```python
set_name.clear()
```

````python
numbers = {10,20,30,40}

numbers.clear()

print(numbers)
````

# copy()

The `copy()` method creates a shallow copy of a Set.

Changes made to the copied Set do not affect the original Set.

````python
A = {1,2,3}

B = A.copy()

B.add(4)

print("Original:", A)
print("Copied:", B)
````

# len()

The `len()` function returns the total number of unique elements present in a Set.

````python
fruits = {"Apple","Banana","Orange","Mango"}

print(len(fruits))
````

````python
numbers = {1,2,2,3,3,4,5}

print(numbers)
print(len(numbers))
````
