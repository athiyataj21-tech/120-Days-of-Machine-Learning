# What is a Set?

A **Set** is one of Python's built-in data structures used to store multiple items in a single variable.

Unlike Lists and Tuples,

- Sets store **only unique values**
- Duplicate values are removed automatically
- Sets are unordered
- Sets are mutable
- Sets are very fast because they use **Hash Tables**

They are mainly used when:

- Removing duplicate values
- Finding common elements
- Performing mathematical operations
- Fast searching

````python
fruits = {"Apple", "Banana", "Orange"}

print(fruits)
````

## Why Use Sets?

Sets are useful in many real-world applications.

Examples:

- Removing duplicate records
- Comparing two datasets
- Finding common friends
- Machine Learning preprocessing
- Recommendation systems
- Database operations

Since Sets are implemented using Hash Tables, searching is much faster than Lists.

````python
numbers = [1,2,2,3,4,5,5,6]

unique_numbers = set(numbers)

print(unique_numbers)
````

# Membership Operators

Python provides two membership operators for Sets.

- `in`
- `not in`

Since Sets use Hash Tables internally, membership testing is extremely fast (average O(1) time complexity).

Membership operators are commonly used to:

- Check whether an element exists
- Validate user input

# Removing Elements

Python provides several methods to remove elements from a Set.

- remove()
- discard()
- pop()
- clear()

Each method behaves differently.

## remove()

The `remove()` method deletes a specified element.

If the element does not exist, Python raises a **KeyError**.

### Syntax

```python
set_name.remove(element)
```

````python
fruits = {"Apple","Banana","Orange"}

fruits.remove("Banana")

print(fruits)
````

````python
fruits = {"Apple","Banana"}

fruits.remove("Mango")
````

## discard()

The `discard()` method also removes an element.

Unlike `remove()`, it does **not** raise an error if the element is absent.

### Syntax

```python
set_name.discard(element)
```

````python
fruits = {"Apple","Banana"}

fruits.discard("Banana")

print(fruits)
````

````python
fruits = {"Apple","Banana"}

fruits.discard("Mango")

print(fruits)
````

# remove() vs discard()

| remove() | discard() |
|-----------|-----------|
| Raises KeyError | No Error |
| Deletes existing element | Deletes existing element |
| Fails if absent | Ignores if absent |

# pop()

The `pop()` method removes and returns a random element from the Set.

Since Sets are unordered, you cannot predict which element will be removed.

````python
colors = {"Red","Green","Blue"}

removed = colors.pop()

print("Removed:", removed)
print("Remaining:", colors)
````

# clear()

The `clear()` method removes all elements from the Set while keeping the Set object itself.

### Syntax

```python
set_name.clear()
```

````python
numbers = {10,20,30,40}

numbers.clear()

print(numbers)
````

# copy()

The `copy()` method creates a shallow copy of a Set.

Changes made to the copied Set do not affect the original Set.

````python
A = {1,2,3}

B = A.copy()

B.add(4)

print("Original:", A)
print("Copied:", B)
````

# Set Operations

One of the biggest advantages of Sets is that they support mathematical set operations.

Python provides built-in methods and operators to perform these operations efficiently.

The most common Set operations are:

- Union
- Intersection
- Difference
- Symmetric Difference

These operations are widely used in:

- Data Analysis
- Machine Learning
- Database Queries
- Recommendation Systems
- Duplicate Removal

# Union

The **Union** of two Sets returns all unique elements from both Sets.

Duplicate values appear only once.

### Syntax

```python
set1.union(set2)
```

or

```python
set1 | set2
```

````python
A = {1, 2, 3}
B = {3, 4, 5}

result = A.union(B)

print(result)
````

````python
A = {"Apple", "Banana"}
B = {"Banana", "Orange"}

print(A | B)
````

## Real-World Example: Union

Suppose two stores have different customer lists.

Using Union, we can find all unique customers who visited either store.

````python
store1 = {"Ali", "John", "David"}
store2 = {"David", "Sara", "Ali", "Maria"}

all_customers = store1.union(store2)

print(all_customers)
````

# Intersection

Intersection returns only the elements that are common to both Sets.

### Syntax

```python
set1.intersection(set2)
```

or

```python
set1 & set2
```

````python
A = {1,2,3,4}
B = {3,4,5,6}

print(A.intersection(B))
````

````python
print(A & B)
````

## Real-World Example: Intersection

Suppose two students enroll in different courses.

The Intersection operation finds students enrolled in both courses.

````python
python_students = {"Asha", "Rahul", "John", "David"}
ml_students = {"David", "John", "Sara"}

common_students = python_students.intersection(ml_students)

print(common_students)
````

# Difference

Difference returns elements that are present in the first Set but not in the second Set.

### Syntax

```python
set1.difference(set2)
```

or

```python
set1 - set2
```

````python
A = {1,2,3,4}
B = {3,4,5}

print(A.difference(B))
````

````python
print(A - B)
````

````python
print(B - A)
````

## Real-World Example: Difference

Suppose a company wants to identify employees who completed Python training but have not yet completed Machine Learning training.

````python
python_course = {"Ali", "Sara", "John", "David"}
ml_course = {"John", "David"}

remaining = python_course - ml_course

print(remaining)
````

# Symmetric Difference

Symmetric Difference returns elements that belong to either Set, but not both.

It removes the common elements.

### Syntax

```python
set1.symmetric_difference(set2)
```

or

```python
set1 ^ set2
```

````python
A = {1,2,3,4}
B = {3,4,5,6}

print(A.symmetric_difference(B))
````

````python
print(A ^ B)
````

## Real-World Example: Symmetric Difference

Suppose two streaming platforms have different movie collections.

The Symmetric Difference identifies movies that are exclusive to one platform.

````python
netflix = {"Movie1", "Movie2", "Movie3"}
prime = {"Movie3", "Movie4", "Movie5"}

exclusive_movies = netflix ^ prime

print(exclusive_movies)
````

# Comparison of Set Operations

| Operation | Method | Operator | Result |
|-----------|--------|----------|--------|
| Union | `union()` | `\|` | All unique elements |
| Intersection | `intersection()` | `&` | Common elements |
| Difference | `difference()` | `-` | Elements in first Set only |
| Symmetric Difference | `symmetric_difference()` | `^` | Non-common elements |

````python
A = {10, 20, 30}
B = {20, 30, 40}

print("Union:", A | B)
print("Intersection:", A & B)
print("Difference:", A - B)
print("Symmetric Difference:", A ^ B)
````

# In-place Set Operations

So far, we have seen methods like:

- union()
- intersection()
- difference()
- symmetric_difference()

These methods return a **new Set** and do not modify the original Set.

Python also provides **in-place methods** that directly modify the existing Set.

These are:

- update()
- intersection_update()
- difference_update()
- symmetric_difference_update()

These methods are memory-efficient because they do not create a new Set.

# update()

The `update()` method adds all unique elements from another iterable into the original Set.

Unlike `union()`, it modifies the original Set.

### Syntax

```python
set.update(iterable)
```

````python
A = {1,2,3}
B = {3,4,5}

A.update(B)

print(A)
````

## update() vs union()

| update() | union() |
|-----------|----------|
| Modifies original Set | Returns new Set |
| No return value | Returns new Set |
| Faster | Slightly slower |

````python
A = {10,20}

B = {20,30}

C = A.union(B)

print("A =",A)
print("C =",C)
````

````python
A = {10,20}

B = {20,30}

A.update(B)

print(A)
````

# intersection_update()

Keeps only the common elements inside the original Set.

### Syntax

```python
set.intersection_update(other_set)
```

````python
A = {1,2,3,4}

B = {3,4,5,6}

A.intersection_update(B)

print(A)
````

# difference_update()

Removes elements present in another Set.

### Syntax

```python
set.difference_update(other_set)
```

````python
A = {1,2,3,4}

B = {3,4}

A.difference_update(B)

print(A)
````

# symmetric_difference_update()

Updates the Set with elements present in either Set but not in both.

### Syntax

```python
set.symmetric_difference_update(other_set)
```

````python
A = {1,2,3}

B = {3,4,5}

A.symmetric_difference_update(B)

print(A)
````

# Superset

A Set is a superset if it contains every element of another Set.


# Disjoint Sets

Python supports Set Comprehensions just like List Comprehensions.

Syntax

````python
squares = {x*x for x in range(10)}

print(squares)
````

# Interview Questions

### Q1 Why are Sets faster than Lists?

Because Sets are implemented using **Hash Tables**.

Average Time Complexity:

- Search → O(1)
- Insert → O(1)
- Delete → O(1)

Whereas Lists require O(n) search.

# Time Complexity

| Operation | Complexity |
|------------|------------|
| Add | O(1) |
| Remove | O(1) |
| Search | O(1) |
| Union | O(len(A)+len(B)) |
| Intersection | O(min(len(A),len(B))) |
| Difference | O(len(A)) |

### Key Takeaways

- Sets store only unique values.
- Sets are unordered and mutable.
- They provide O(1) average-time complexity for search, insertion, and deletion.
- Sets are widely used in Machine Learning, Data Science, NLP, database systems, and recommendation engines for tasks such as duplicate removal, membership testing, and set operations.
