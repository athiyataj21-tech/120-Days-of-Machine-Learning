## Creating Strings

Use single, double, or triple quotes.

```python
s1="Hello"
s2='World'
s3="""Python"""
print(s1,s2,s3)
```

## Indexing

Positive and negative indexing.

```python
text="Machine Learning"
print(text[0])
print(text[-1])
```

## Slicing

start:stop:step examples.

```python
text="Machine Learning"
print(text[:7])
print(text[8:])
print(text[::-1])
```

## Operations

Concatenation, repetition, membership.

```python
a="Machine"
b="Learning"
print(a+" "+b)
print("ML"*3)
print("Learn" in b)
```

## Methods

Common methods: upper, lower, strip, split, join, replace, find, count.

```python
text="  hello python  "
print(text.upper())
print(text.strip())
print("a,b,c".split(","))
print("-".join(["ML","AI"]))
```

## Formatting

f-strings, format(), % formatting.

```python
name="Athiya"
day=10
print(f"Hello {name}, Day {day}")
```

## Escape Characters

\n, \t, \\, quotes.

```python
print("Hello\nWorld")
print("A\tB")
```

## Part 2 – String Operations & Methods

## String Operations

```python
a="Machine"; b="Learning"
print(a+" "+b)
print("AI "*3)
print("Machine" in "Machine Learning")
```

## Built-in Functions

```python
text="Python"
print(len(text), max(text), min(text))
print(sorted(text))
print(list(reversed(text)))
for i,c in enumerate(text): print(i,c)
print(ord("A"), chr(65))
```

## String Methods

```python
text=" machine learning "
print(text.upper())
print(text.title())
print(text.strip())
print(text.replace("machine","deep"))
print("banana".count("a"))
print("Machine".find("h"))
```

## Validation Methods

```python
print("Python".isalpha())
print("123".isdigit())
print("Python123".isalnum())
print(" ".isspace())
```
