# 1. Arithmetic Operators

Arithmetic operators are used to perform mathematical calculations.

```python
a = 15
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
```

## Output Explanation

Python supports all common mathematical operations.

Floor division removes the decimal part.

Exponent is used for powers.

# 2. Assignment Operators

```python
x = 10

print("Initial value:", x)

x += 5
print("After += :", x)

x -= 2
print("After -= :", x)

x *= 3
print("After *= :", x)

x /= 2
print("After /= :", x)

x //= 2
print("After //= :", x)

x %= 3
print("After %= :", x)
```

# 3. Comparison Operators

Comparison operators always return either

True

or

False

```python
a = 10
b = 20

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

# 4. Logical Operators

```python
age = 22
salary = 40000

print(age > 18 and salary > 30000)
print(age > 25 or salary > 30000)
print(not age > 18)
```

# 5. Identity Operators

Identity operators compare memory locations.

```python
list1 = [1,2,3]
list2 = list1
list3 = [1,2,3]

print(list1 is list2)
print(list1 is list3)

print(list1 == list3)
```

## Difference

== compares values.

is compares memory addresses.

# 6. Membership Operators

```python
fruits = ["apple","banana","orange"]

print("apple" in fruits)
print("grapes" in fruits)

print("grapes" not in fruits)
```

# 7. Bitwise Operators

```python
a = 5
b = 3

print("AND :", a & b)
print("OR :", a | b)
print("XOR :", a ^ b)
print("NOT :", ~a)
print("Left Shift :", a << 1)
print("Right Shift :", a >> 1)
```

# 8. Operator Precedence

Python follows this order

1. ()
2. **
3. *, /, //, %
4. +, -
5. Comparison
6. not
7. and
8. or

```python
result = 5 + 2 * 3 ** 2

print(result)
```

# 9. Machine Learning Example

Operators are heavily used while evaluating models.

```python
accuracy = 94
precision = 91

if accuracy >= 90 and precision >= 90:
    print("Excellent Model")
else:
    print("Needs Improvement")
```
