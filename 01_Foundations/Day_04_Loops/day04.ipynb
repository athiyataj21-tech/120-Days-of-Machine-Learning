## for Loop

A `for` loop is used to iterate over a sequence like a list, tuple, string, or range.

````python
fruits = ["Apple", "Banana", "Mango"]

for fruit in fruits:
    print(fruit)
````

````python
name = "Python"

for letter in name:
    print(letter)
````

## range() Function

The `range()` function generates a sequence of numbers.

Syntax:

```python
range(start, stop, step)
```

````python
for i in range(5):
    print(i)
````

````python
for i in range(2, 11, 2):
    print(i)
````

## while Loop

A `while` loop executes as long as the condition is True.

````python
count = 1

while count <= 5:
    print(count)
    count += 1
````

## break Statement

The `break` statement immediately terminates the loop.

````python
for i in range(1, 11):

    if i == 6:
        break

    print(i)
````

## continue Statement

The `continue` statement skips the current iteration.

````python
for i in range(1, 11):

    if i % 2 == 0:
        continue

    print(i)
````

## pass Statement

The `pass` statement is a placeholder and does nothing.

````python
for i in range(5):
    pass

print("Loop Completed")
````

## Nested Loops

A loop inside another loop is called a nested loop.

````python
for i in range(1, 4):

    for j in range(1, 4):
        print(i, j)
````

## Pattern Printing

````python
for i in range(5):

    for j in range(i + 1):
        print("*", end=" ")

    print()
````

````python
for i in range(5, 0, -1):

    for j in range(i):
        print("*", end=" ")

    print()
````

## Multiplication Table

````python
num = 7

for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
````

## Sum of First 10 Numbers

````python
total = 0

for i in range(1, 11):
    total += i

print("Sum =", total)
````

## Factorial

````python
number = 5
factorial = 1

for i in range(1, number + 1):
    factorial *= i

print("Factorial =", factorial)
````
